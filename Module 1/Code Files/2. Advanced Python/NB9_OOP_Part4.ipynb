```python
# Abstract Method: 
#  Sometimes we don't know about implementation, still we can declare a method. Such 
# types of methods are called abstract methods.i.e abstract method has only declaration 
# but not implementation. 
#  In python we can declare abstract method by using @abstractmethod decorator as 
# follows. 
 
#  @abstractmethod 
#  def m1(self): pass 
 
#  @abstractmethod decorator present in abc module. Hence compulsory we should 
# import abc module,otherwise we will get error. 
#  abc  abstract base class module 
class Test:   
    @abstractmethod   
    def m1(self):   
        pass 
```

```python
# # Abstract class: 
# # Some times implementation of a class is not complete,such type of partially 
# # implementation classes are called abstract classes. Every abstract class in Python should 
# # be derived from ABC class which is present in abc module. 


# Case 1:

# In this case, you’re simply creating a normal class and instantiating it, which works because it’s a concrete class.

# Case 1
from abc import *

class Test:
    pass

t = Test()  # This will work because 'Test' is a concrete class

# In the above code we can create object for Test class b'z it is concrete class and it does not 
# conatin any abstract method.


# Case 2:

# Here, the class Test inherits from ABC, but since it doesn’t contain any abstract methods, it can be instantiated.


# Case 2
from abc import *

class Test(ABC):
    pass

t = Test()  # This will work because there are no abstract methods

# In the above code we can create object, even it is derived from ABC class,b'z it does not 
# contain any abstract method.



#Case 3:

#This case defines an abstract class with an abstract method. You cannot instantiate this class directly, and you must either implement the m1() method in a subclass or instantiate a subclass that implements it.


# Case 3
from abc import ABC, abstractmethod

class Test(ABC):
    @abstractmethod
    def m1(self):
        pass

# The following line will raise an error if uncommented
# t = Test()  # This will raise an error because 'Test' is an abstract class and cannot be instantiated

# To instantiate, you would need a subclass that implements 'm1'
class SubTest(Test):
    def m1(self):
        print("m1 method implemented")

# Now, you can instantiate the subclass
t = SubTest()
t.m1()  # This will work, and 'm1' will print the message

    
```

```python
from abc import ABC, abstractmethod

# Abstract base class
class Vehicle(ABC):
    @abstractmethod
    def no_of_wheels(self):
        pass

# Concrete subclass that implements the abstract method
class Bus(Vehicle):
    def no_of_wheels(self):
        return 7

# Another concrete subclass that implements the abstract method
class Auto(Vehicle):
    def no_of_wheels(self):
        return 3

# Creating object of Bus and calling the method
b = Bus()
print(b.no_of_wheels())  # Output: 7

# Creating object of Auto and calling the method
a = Auto()
print(a.no_of_wheels())  # Output: 3
```

```python
# Interfaces In Python: 
# In general if an abstract class contains only abstract methods such type of abstract class is 
# considered as interface. 


from abc import ABC, abstractmethod

# Abstract base class for database interfaces
class DBInterface(ABC):
    @abstractmethod
    def connect(self):
        pass

    @abstractmethod
    def disconnect(self):
        pass

# Concrete class for Oracle database
class Oracle(DBInterface):
    def connect(self):
        print('Connecting to Oracle Database...')

    def disconnect(self):
        print('Disconnecting from Oracle Database...')

# Concrete class for Sybase database
class Sybase(DBInterface):
    def connect(self):
        print('Connecting to Sybase Database...')

    def disconnect(self):
        print('Disconnecting from Sybase Database...')

# Main logic
if __name__ == "__main__":
    dbname = input("Enter Database Name (Oracle/Sybase): ")
    
    try:
        # Dynamically get the class from the globals() dictionary
        classname = globals()[dbname]
        if issubclass(classname, DBInterface):
            x = classname()
            x.connect()
            x.disconnect()
        else:
            print("Invalid database class. It does not implement DBInterface.")
    except KeyError:
        print("Error: No such class found. Please enter 'Oracle' or 'Sybase'.")
```

```python
# Concreate class vs Abstract Class vs Inteface: 
# 1) If we dont know anything about implementation just we have requirement 
# specification then we should go for interface. 
# 2) If we are talking about implementation but not completely then we should go for 
# abstract class. (partially implemented class). 
# 3) If we are talking about implementation completely and ready to provide service then 
# we should go for concrete class. 
```

```python
from abc import ABC, abstractmethod

# Abstract base class
class CollegeAutomation(ABC):
    @abstractmethod
    def m1(self):
        pass

    @abstractmethod
    def m2(self):
        pass

    @abstractmethod
    def m3(self):
        pass

# Partially implemented abstract class
class AbsCls(CollegeAutomation):
    def m1(self):
        print('m1 method implementation')

    def m2(self):
        print('m2 method implementation')

# Fully implemented concrete class
class ConcreteCls(AbsCls):
    def m3(self):
        print('m3 method implementation')

# Instantiate and call methods
c = ConcreteCls()
c.m1()
c.m2()
c.m3()
```

```python
# Public, Protected and Private Attributes: 
 
# By default every attribute is public. We can access from anywhere either within the class 
# or from outside of the class. 
# Eg: name = 'durga' 
 
# Protected attributes can be accessed within the class anywhere but from outside of the 
# class only in child classes. We can specify an attribute as protected by prefexing with _ 
# symbol. 
 
# Syntax:  _variablename = value 
# Eg:  _name='durga'
# But is is just convention and in reality does not exists protected attributes. 
 
# private attributes can be accessed only within the class.i.e from outside of the class we 
# cannot access. We can declare a variable as private explicitly by prefexing with 2 
# underscore symbols. 
 
# syntax:   __variablename=value



class Test:
    x = 10        # Public variable
    _y = 20       # Protected variable (by convention)
    __z = 30      # Private variable (name mangled)

    def m1(self):
        print(Test.x)       # Accessing public
        print(Test._y)      # Accessing protected
        print(Test.__z)     # Accessing private (within class is allowed)

# Create object
t = Test()
t.m1()

# Accessing from outside the class
print(Test.x)        # OK: public
print(Test._y)       # OK: protected (by convention, still accessible)

# This line will raise AttributeError due to name mangling
# print(Test.__z)    # ❌ Will raise AttributeError

# Correct way to access the name-mangled variable __z
print(Test._Test__z)  # ✅ Accessing the private variable using name mangling
```

```python
# __str__() method: 
 
#  Whenever we are printing any object reference internally __str__() method will be 
# called which is returns string in the following format                                                         
# <__main__.classname object at 0x022144B0> 
 
#  To return meaningful string representation we have to override __str__() method.



class Student:
    def __init__(self, name, rollno):
        self.name = name
        self.rollno = rollno

    def __str__(self):
        return 'This is Student with Name: {} and Rollno: {}'.format(self.name, self.rollno)

# Creating student objects
s1 = Student('Durga', 101)
s2 = Student('Ravi', 102)

# Printing student details
print(s1)
print(s2)
```

```python
# Output without Overriding str(): 
# <__main__.Student object at 0x022144B0> 
# <__main__.Student object at 0x022144D0> 
 
# Output with Overriding str(): 
# This is Student with Name: Durga and Rollno: 101 
# This is Student with Name: Ravi and Rollno: 102 
 
# Difference between str() and repr() 
# OR 
# Difference between __str__() and __repr__() 
#  str() internally calls __str__() function and hence functionality of both is same. 
#  Similarly,repr() internally calls __repr__() function and hence functionality of both is 
# same. 
#  str() returns a string containing a nicely printable representation object. 
#  The main purpose of str() is for readability.It may not possible to convert result string 
# to original object. 
 
 
```

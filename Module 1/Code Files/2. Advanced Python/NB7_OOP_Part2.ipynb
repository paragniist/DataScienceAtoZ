```python
# # Using Members of One Class inside Another Class: 
# # We can use members of one class inside another class by using the following ways 
 
# # 1) By Composition (Has-A Relationship) 
# # 2) By Inheritance (IS-A Relationship)


# What is Composition (Has-A Relationship)?

# Composition means building a class by including objects of other classes inside it.

# It models a “Has-A” relationship (one class has a reference to another class).

# Example:

# A Car has a Engine.

# A Person has a Address.





class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print(f"Engine with {self.horsepower} HP started.")

class Car:
    def __init__(self, brand, horsepower):
        self.brand = brand
        # Car HAS-A Engine (composition)
        self.engine = Engine(horsepower)

    def drive(self):
        print(f"{self.brand} car is ready to drive.")
        self.engine.start()

# Create Car object (Car HAS-A Engine)
c1 = Car("Toyota", 150)
c1.drive()
```

```python
# By Inheritance (IS-A Relationship): 
# What ever variables, methods and constructors available in the parent class by default 
# available to the child classes and we are not required to rewrite. Hence the main 
# advantage of inheritance is Code Reusability and we can extend existing functionality 
# with some more extra functionality

# If class B inherits from class A, we say B is-a type of A.

# This is used to model generalization → specialization.
# Example: Dog is a Animal, Car is a Vehicle.




# Parent class
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print(f"Brand: {self.brand}")

# Child class (inherits from Vehicle)
class Car(Vehicle):   # Car is-a Vehicle
    def __init__(self, brand, model):
        super().__init__(brand)   # call parent constructor
        self.model = model

    def show_details(self):
        print(f"Car Brand: {self.brand}, Model: {self.model}")

# Another child class (inherits from Vehicle)
class Bike(Vehicle):   # Bike is-a Vehicle
    def __init__(self, brand, cc):
        super().__init__(brand)
        self.cc = cc

    def show_details(self):
        print(f"Bike Brand: {self.brand}, Engine: {self.cc}cc")

# Usage
c = Car("Toyota", "Corolla")
c.show_brand()        # Inherited method
c.show_details()      # Car-specific method

b = Bike("Yamaha", 150)
b.show_brand()        # Inherited method
b.show_details()      # Bike-specific method
```

```python
# What ever methods present in Parent class are automatically available to the child class 
# and hence on the child class reference we can call both parent class methods and child 
# class methods. 



# Parent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat_and_drink(self):
        print("Eat Biryani and Drink Beer")


# Child class (Employee IS-A Person)
class Employee(Person):
    def __init__(self, name, age, eno, esal):
        super().__init__(name, age)   # Call parent constructor
        self.eno = eno
        self.esal = esal

    def work(self):
        print("Coding Python is very easy, just like drinking chilled Beer")

    def emp_info(self):
        print("Employee Name:", self.name)
        print("Employee Age:", self.age)
        print("Employee Number:", self.eno)
        print("Employee Salary:", self.esal)


# Usage
e = Employee("Durga", 48, 100, 10000)
e.eat_and_drink()   # Inherited method from Person
e.work()            # Employee-specific method
e.emp_info()        # Employee-specific method
```

```python
# IS-A vs HAS-A Relationship: 
#  If we want to extend existing functionality with some more extra functionality then we 
# should go for IS-A Relationship. 
#  If we dont want to extend and just we have to use existing functionality then we 
# should go for HAS-A Relationship. 
#  Eg:  Employee class extends Person class Functionality But Employee class just uses Car 
# functionality but not extending




# IS-A Relationship (Inheritance)

# Meaning: One class is a type of another class.

# Implemented using inheritance (class Child(Parent): ...).

# It represents a generalization → specialization relationship.


# IS-A Relationship (Inheritance)
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):  # Employee IS-A Person
    def __init__(self, name, eno):
        super().__init__(name)
        self.eno = eno

# Usage
e = Employee("Durga", 100)
print(isinstance(e, Person))   # True → Employee IS-A Person








#HAS-A Relationship (Composition / Aggregation)

# Meaning: One class contains an object of another class.

# Implemented using object composition (creating an instance of one class inside another).

# It represents a part-of or ownership relationship.



# HAS-A Relationship (Composition)
class Car:
    def __init__(self, brand):
        self.brand = brand

class Employee:
    def __init__(self, name, car):
        self.name = name
        self.car = car   # Employee HAS-A Car

    def show_info(self):
        print(f"Employee: {self.name}, Car: {self.car.brand}")

# Usage
c = Car("Toyota")
e = Employee("Durga", c)
e.show_info()
```

```python
# Class representing a Car (used in HAS-A relationship)
class Car:
    def __init__(self, name, model, color):
        self.name = name
        self.model = model
        self.color = color

    def get_info(self):
        print(f"\tCar Name: {self.name}\n\tModel: {self.model}\n\tColor: {self.color}")


# Parent class (Person)
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat_and_drink(self):
        print("Eat Biryani and Drink Beer")


# Child class (Employee IS-A Person and HAS-A Car)
class Employee(Person):
    def __init__(self, name, age, eno, esal, car):
        super().__init__(name, age)   # Inherit from Person
        self.eno = eno
        self.esal = esal
        self.car = car               # Composition: Employee HAS-A Car

    def work(self):
        print("Coding Python is very easy, just like drinking chilled Beer")

    def emp_info(self):
        print("Employee Name:", self.name)
        print("Employee Age:", self.age)
        print("Employee Number:", self.eno)
        print("Employee Salary:", self.esal)
        print("Employee Car Info:")
        self.car.get_info()


# Usage
c = Car("Innova", "2.5V", "Grey")
e = Employee("Durga", 48, 100, 10000, c)

# Methods
e.eat_and_drink()   # From Person
e.work()            # From Employee
e.emp_info()        # Shows Employee details + Car info
```

```python
# Composition vs Aggregation: 
 
# Composition: 
# Without existing container object if there is no chance of existing contained object then 
# the container and contained objects are strongly associated and that strong association is 
# nothing but Composition. 
  
# Eg: University contains several Departments and without existing university object there 
# is no chance of existing Department object. Hence University and Department objects are 
# strongly associated and this strong association is nothing but Composition. 





# Aggregation: 
# Without existing container object if there is a chance of existing contained object then the 
# container and contained objects are weakly associated and that weak association is 
# nothing but Aggregation. 
 
# Eg: Department contains several Professors. Without existing Department still there may 
# be a chance of existing Professor. Hence Department and Professor Objects are weakly 
# associated, which is nothing but Aggregation. 
```

```python
#Coding Example: 
class Student:   
    collegeName='DURGASOFT'   
    def __init__(self,name):   
        self.name=name   
print(Student.collegeName)   
s=Student('Durga')   
print(s.name)   
#Output: 
#DURGASOFT 
# Durga 
 
# In the above example without existing Student object there is no chance of existing his 
# name. Hence Student Object and his name are strongly associated which is nothing but 
# Composition. 
 
# But without existing Student object there may be a chance of existing collegeName. Hence 
# Student object and collegeName are weakly associated which is nothing but Aggregation. 
# : 
: 
# Conclusion: 
# The relation between object and its instance variables is always Composition where as the 
# relation between object and static variables is Aggregation. 
 
# Note: Whenever we are creating child class object then child class constructor will be 
# executed. If the child class does not contain constructor then parent class constructor will 
# be executed, but parent object won't be created. 
```

```python
# Types of Inheritance: 

# Single Inheritance: 
# The concept of inheriting the properties from one class to another class is known as 
# single inheritance. 
class P:   
 def m1(self):   
     print("Parent Method")   
class C(P):   
     def m2(self):   
       print("Child Method")   
c=C()   
c.m1()   
c.m2()  
```

```python
#  Multi Level Inheritance: 
# The concept of inheriting the properties from multiple classes to single class with the 
# concept of one after another is known as multilevel inheritance. 
class P:  
    def m1(self):   
        print("Parent Method")   
class C(P):   
    def m2(self):   
        print("Child Method")   
class CC(C):   
    def m3(self):   
        print("Sub Child Method")   
c=CC()
c.m1()   
c.m2()   
c.m3()  
```

```python
# ) Hierarchical Inheritance: 
# The concept of inheriting properties from one class into multiple classes which are 
# present at same level is known as Hierarchical Inheritance



# Parent class
class P:
    def m1(self):
        print("Parent Method")


# Child class 1 (inherits from P)
class C1(P):
    def m2(self):
        print("Child1 Method")


# Child class 2 (inherits from P)
class C2(P):
    def m3(self):
        print("Child2 Method")


# Usage
c1 = C1()
c1.m1()   # Inherited from P
c1.m2()   # From C1

c2 = C2()
c2.m1()   # Inherited from P
c2.m3()   # From C2
```

```python
#  Multiple Inheritance: 
# The concept of inheriting the properties from multiple classes into a single class at a 
# time, is known as multiple inheritance.



# Parent class 1
class P1:
    def m1(self):
        print("Parent1 Method")


# Parent class 2
class P2:
    def m2(self):
        print("Parent2 Method")


# Child class (inherits from P1 and P2 → Multiple Inheritance)
class C(P1, P2):
    def m3(self):
        print("Child Method")


# Usage
c = C()
c.m1()   # From P1
c.m2()   # From P2
c.m3()   # From C
```

```python
# Method Resolution Order (MRO): 
#  In Hybrid Inheritance the method resolution order is decided based on MRO 
# algorithm. 
#  This algorithm is also known as C3 algorithm. 
#  Samuele Pedroni proposed this algorithm. 
#  It follows DLR (Depth First Left to Right) i.e Child will get more priority than Parent.  
#  Left Parent will get more priority than Right Parent. 
#  MRO(X) = X+Merge(MRO(P1),MRO(P2),...,ParentList) 
 
# Head Element vs Tail Terminology: 
#  Assume C1,C2,C3,...are classes. 
#  In the list: C1C2C3C4C5.... 
#  C1 is considered as Head Element and remaining is considered as Tail. 
```

```python
# How to find Merge: 
#  Take the head of first list 
#  If the head is not in the tail part of any other list, then add this head to the result and 
# remove it from the lists in the merge. 
#  If the head is present in the tail part of any other list, then consider the head element 
# of the next list and continue the same process. 
 
# Note: We can find MRO of any class by using mro() function. 
#                print(ClassName.mro())
```

```python
# super() Method: 
# super() is a built-in method which is useful to call the super class constructors,variables 
# and methods from the child class



# Parent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Child class (inherits from Person)
class Student(Person):
     def __init__(self, name, age, rollno, marks):
        # Call parent constructor
        super().__init__(name, age)
        self.rollno = rollno
        self.marks = marks

    # Overriding display() method
     def display(self):
        super().display()   # Call parent display()
        print("Roll No:", self.rollno)
        print("Marks:", self.marks)


# Usage
s1 = Student("Durga", 22, 101, 90)
s1.display()
```

```python
# Various Important Points about super(): 
 
# Case-1: From child class we are not allowed to access parent class instance variables by 
# using super(), Compulsory we should use self only. 
# But we can access parent class static variables by using super(). 



# Parent class
# Parent class
class P:
    a = 10   # Class variable

    def __init__(self):
        self.b = 20   # Instance variable


# Child class
class C(P):
    def m1(self):
        # Accessing parent class variable using super()
        print("Accessing parent class variable a with super():", super().a)   # ✅ Valid

        # Accessing instance variable with self
        print("Accessing instance variable b with self:", self.b)             # ✅ Valid

        # Trying to access instance variable with super()
        # This is invalid because super() only gives access to methods and class variables, not instance vars
        # print(super().b)   # ❌ Invalid, will raise AttributeError


# Usage
c = C()
c.m1()
```

```python
# Case-2: From child class constructor and instance method, we can access parent class 
# instance method, static method and class method by using super()


# Parent class
class P:
    def __init__(self):
        print("Parent Constructor")

    def m1(self):
        print("Parent Instance Method")

    @classmethod
    def m2(cls):
        print("Parent Class Method")

    @staticmethod
    def m3():
        print("Parent Static Method")


# Child class
class C(P):
    def __init__(self):
        # Calling parent constructor and methods from constructor
        super().__init__()   # Calls parent constructor
        super().m1()         # Calls parent instance method
        super().m2()         # Calls parent class method
        super().m3()         # Calls parent static method

    def m1(self):
        # Calling parent methods from overridden method
        super().__init__()   # Calls parent constructor
        super().m1()         # Calls parent instance method
        super().m2()         # Calls parent class method
        super().m3()         # Calls parent static method


# Usage
c = C()      # Triggers constructor of C
c.m1()       # Calls overridden m1() in C
```

```python
# Case-3: From child class, class method we cannot access parent class instance methods 
# and constructors by using super() directly(but indirectly possible). But we can access 
# parent class static and class methods.


# Parent class
class P:
    def __init__(self):
        print("Parent Constructor")

    def m1(self):
        print("Parent Instance Method")

    @classmethod
    def m2(cls):
        print("Parent Class Method")

    @staticmethod
    def m3():
        print("Parent Static Method")


# Child class
class C(P):
    @classmethod
    def m1(cls):
        # super().__init__()   # ❌ Invalid (constructor is an instance method)
        # super().m1()         # ❌ Invalid (instance method cannot be called in classmethod)

        # ✅ Valid: calling parent class method & static method
        super().m2()
        super().m3()


# Usage
C.m1()
```

```python
# From Class Method of Child Class, how to call Parent Class Instance Methods 
# and Constructors:



class A:
    def __init__(self):
        print("Parent constructor")

    def m1(self):
        print("Parent instance method")


class B(A):
    @classmethod
    def m2(cls):
        # To call parent instance methods, first create an object
        obj = cls()          # this will call A.__init__()
        obj.m1()             # calling parent instance method


# Usage
B.m2()
```

```python
# Case-4: In child class static method we are not allowed to use super() generally (But in 
# special way we can use



class P:
    def __init__(self):
        print("Parent Constructor")

    def m1(self):
        print("Parent instance method")

    @classmethod
    def m2(cls):
        print("Parent class method")

    @staticmethod
    def m3():
        print("Parent static method")


class C(P):
    @staticmethod
    def m1():
        # We cannot use super() in static methods, so call directly using class name
        P().m1()     # create object to call instance method
        P.m2()       # call parent class method directly
        P.m3()       # call parent static method directly


# Usage
C.m1()
```
